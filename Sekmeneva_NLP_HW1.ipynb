````python
import string
````

````python
# Функция для очистки текста от знаков препинания
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
````

````python
# Используем функцию FreqDist из nltk для подсчета частоты встречаемости слов
fdist = FreqDist()
all_words = []
# Перебор файлов в каталоге с оценкой скорости выполнения
for filename in tqdm(os.listdir(directory_path), desc="Processing Files"):
    with open(os.path.join(directory_path, filename), 'r') as file:
        text = file.read()
        # Очистка текста от пунктуации
        text = remove_punctuation(text)

        words = text.split()

        # Фильтрация стоп-слов и обновление частотного распределения
        filtered_words = [word for word in words if word not in stop_words]
        fdist.update(filtered_words)


# Вывод топ-1000 слов
top_words = fdist.most_common(1000)

for word, frequency in top_words:
    print(f'{word}: {frequency}')
````

````python
# Загрузка более крупной предобученной модели spaCy
nlp = spacy.load("en_core_web_lg")

# Так как в файлах есть тексты превышающие стандартные настройки SpaCy, то сразу их переопределим
nlp.max_length = 2000000
````

````python
# Инициализация пустых словарей для подсчета имен и имен+фамилий, а также сочетаний "professor" + (имя и/или фамилия)
first_names_count = {}
full_names_count = {}
professor_combinations_count = {}
````

````python
import multiprocessing
from multiprocessing import Pool
````

````python
# Инициализация пустых словарей для подсчета имен и имен+фамилий, а также сочетаний "professor" + (имя и/или фамилия)
first_names_count = multiprocessing.Manager().dict()
full_names_count = multiprocessing.Manager().dict()
professor_combinations_count = multiprocessing.Manager().dict()
````

````python
import datetime
````

````python
def process_file(filename, directory_path):
    file = open(os.path.join(directory_path, filename), 'r')
    text = file.read()
    doc = nlp(text)
    dt_now = datetime.datetime.now()
    print(dt_now)
    for ent in doc.ents:
        if ent.label_ == 'PERSON':
            # Разделяем текст сущности по пробелам
            words = ent.text.split()

            # Проверяем, что сущность состоит из двух слов
            if len(words) == 2:
                # Объединяем имя и фамилию в одну строку
                full_name = ' '.join(words)

                # Добавляем полное имя в словарь и увеличиваем его счетчик
                if full_name in full_names_count:
                    full_names_count[full_name] += 1
                else:
                    full_names_count[full_name] = 1

                # Добавляем имя (первую часть) в словарь и увеличиваем его счетчик
                first_name = words[0]
                if first_name in first_names_count:
                    first_names_count[first_name] += 1
                else:
                    first_names_count[first_name] = 1

        # Проверяем наличие сочетания "professor" в тексте сущности
        if 'professor' in ent.text.lower():
            # Добавляем сочетание "professor" + имя и/или фамилия в словарь и увеличиваем его счетчик
            if ent.text in professor_combinations_count:
                professor_combinations_count[ent.text] += 1
            else:
                professor_combinations_count[ent.text] = 1
````

````python
# Создаем пул процессов
pool = multiprocessing.Pool(32)

# Список файлов в директории
files = os.listdir(directory_path)[:5000]
````

````python
# Обработка файлов в параллельных процессах
for _ in pool.starmap(process_file, [(filename, directory_path) for filename in files]):
    pass
````

````python
# Получаем топ-10
top_first_names = sorted(first_names_count.items(), key=lambda x: x[1], reverse=True)[:10]
top_full_names = sorted(full_names_count.items(), key=lambda x: x[1], reverse=True)[:10]
top_professor_combinations = sorted(professor_combinations_count.items(), key=lambda x: x[1], reverse=True)[:10]

# Выводим результаты
print("Топ-10 имен:")
for name, count in top_first_names:
    print(f"{name}: {count}")

print("\nТоп-10 имен+фамилий:")
for name, count in top_full_names:
    print(f"{name}: {count}")

print("\nТоп-10 сочетаний 'professor' + (имя и/или фамилия):")
for combination, count in top_professor_combinations:
    print(f"{combination}: {count}")
````

````python
import os
import re
from collections import Counter
````

````python
# Путь к фанфикам
directory_path_fanfic = '/content/fanfiction_texts'

# Шаблоны для поиска нужных мотивов
name_pattern = re.compile(r'\b[A-Z][a-zA-Z]*\b')
name_sur_pattern = re.compile(r'\b[A-Z][a-zA-Z]*\s[A-Z][a-zA-Z]*\b')
professor_pattern = re.compile(r'\bprofessor (?:[A-Z][a-zA-Z]*|[A-Z][a-zA-Z]* [A-Z][a-zA-Z]*)\b', re.IGNORECASE)
````

````python
# Функция для поиска имен
def find_names(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
        names = name_pattern.findall(text)
        return names

# Функция для поиска имен и фамилий
def find_names_and_surnames(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
        names_and_surnames = name_sur_pattern.findall(text)
        return names_and_surnames

# Функция для поиска сочетаний
def find_professor_names_in_file(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
        matches = professor_pattern.findall(text)
        return matches
````

````python
# Пройдемся по всем файлам в директории и найдем имена
found_names = []
found_names_and_surnames = []
found_professor_names = []

for root, dirs, files in os.walk(directory_path):
    for file in tqdm(files):
        file_path = os.path.join(root, file)
        names = find_names(file_path)
        found_names.extend(names)

        names_and_surnames = find_names_and_surnames(file_path)
        found_names_and_surnames.extend(names_and_surnames)

        professors = find_professor_names_in_file(file_path)
        found_professor_names.extend(professors)


# Исключим стоп-слова из списка найденных имен
filtered_names = [name for name in found_names if name.lower() not in stop_words]
filtered_names_and_surnames = [name for name in found_names_and_surnames if all(word.lower() not in stop_words for word in name.split())]

# Подсчитаем частоту встречаемости имен
name_counter = Counter(filtered_names)
name_and_surname_counter = Counter(filtered_names_and_surnames)
professor_name_counter = Counter(found_professor_names)

# Получим топ-10 наиболее часто встречающихся имен
top_names = name_counter.most_common(10)
top_name_and_surname = name_and_surname_counter.most_common(10)
top_professor_names = professor_name_counter.most_common(10)
````

````python
# Выведем топ-10
print("Топ-10 имен:")
for name, count in top_names:
    print(f"{name}: {count}")

print("\nТоп-10 имен и фамилий:")
for name_and_surname, count in top_name_and_surname:
    print(f"{name_and_surname}: {count} ")

print("\nТоп-10 имен и фамилий профессоров:")
for professor_name, count in top_professor_names:
  print(f'{professor_name}: {count}')
````

````python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
````

````python
# Загрузка данных
# Объединяем все тексты в один список
all_texts = train_data['text'].tolist() + dev_data['text'].tolist() + test_data['text'].tolist()
````

````python
# Создаем и обучаем токенизатор
tokenizer = Tokenizer()
tokenizer.fit_on_texts(all_texts)
````

````python
# Преобразование текста в последовательности чисел
train_sequences = tokenizer.texts_to_sequences(train_data['text'])
dev_sequences = tokenizer.texts_to_sequences(dev_data['text'])
test_sequences = tokenizer.texts_to_sequences(test_data['text'])
````

````python
# Дополняем последовательности до одинаковой длины
max_sequence_length = max(len(seq) for seq in train_sequences + dev_sequences + test_sequences)
train_sequences = pad_sequences(train_sequences, maxlen=max_sequence_length, padding='post')
dev_sequences = pad_sequences(dev_sequences, maxlen=max_sequence_length, padding='post')
test_sequences = pad_sequences(test_sequences, maxlen=max_sequence_length, padding='post')
````

````python
# Кодируем метки классов
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_data['label'])
dev_labels_encoded = label_encoder.transform(dev_data['label'])
test_labels_encoded = label_encoder.transform(test_data['label'])

num_classes = len(label_encoder.classes_)
````

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout
````

````python
# Создаем модель CNN
# Первая итерация. Результат: Переобучение, и macro-F1 0.07
model_CNN = Sequential()
model_CNN.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=100, input_length=max_sequence_length))
model_CNN.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model_CNN.add(GlobalMaxPooling1D())
model_CNN.add(Dense(64, activation='relu'))
model_CNN.add(Dropout(0.5))
model_CNN.add(Dense(num_classes, activation='softmax'))
````

````python
# Оценка на данных для отладки
dev_predictions = model_CNN.predict(dev_sequences)
dev_labels_predicted = [tf.argmax(label).numpy() for label in dev_predictions]
print("Отчет о классификации на данных для отладки:")
print(classification_report(dev_labels_encoded, dev_labels_predicted))

# Оценка на тестовых данных
test_predictions = model_CNN.predict(test_sequences)
test_labels_predicted = [tf.argmax(label).numpy() for label in test_predictions]
print("Отчет о классификации на тестовых данных:")
print(classification_report(test_labels_encoded, test_labels_predicted))
````

````python
# Компилируем модель
model_CNN.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Обучаем модель
model_CNN.fit(train_sequences, train_labels_encoded, epochs=10, batch_size=64, validation_data=(dev_sequences, dev_labels_encoded))
````

````python
import nltk
from nltk.corpus import wordnet
import random
````

````python
# Функция для получения синонима слова из WordNet
def get_synonym(word):
    synonyms = []
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.append(lemma.name())
    return synonyms
````

````python
# Количество аугментаций для каждого текста
num_augmentations = 3

# Аугментация данных синонимами
augmented_data = []

for index, row in train_data.iterrows():
    text = row['text']
    label = row['label']  # Сохраняем лейбл из исходного датафрейма
    augmented_data.append({'text': text, 'label': label})  # Добавляем исходный текст и лейбл
    words = text.split()
    for _ in range(num_augmentations):
        augmented_text = []
        for word in words:
            synonyms = get_synonym(word)
            if synonyms:
                synonym = random.choice(synonyms)
                augmented_text.append(synonym)
            else:
                augmented_text.append(word)
        augmented_data.append({'text': " ".join(augmented_text), 'label': label})  # Добавляем аугментированный текст и лейбл

# Создаем новый датафрейм с аугментированными данными
augmented_df = pd.DataFrame(augmented_data)
````
